```python
#Importando as bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use("seaborn")
```

```python
#Criando nosso DataFrame
df = pd.read_excel("AdventureWorks.xlsx")
```

```python
#visualizando as 5 primeiras linhas
df.head(5)
```

```python
#Quantidade de linhas e colunas
df.shape
```

```python
#Verificando os tipos de dados
df.dtypes
```

```python
#Qual a receita total?
df["Valor Venda"].sum()
```

```python
#Criando a coluna de custo
df["Custo"] = df["Custo Unitário"].mul(df["Quantidade"]) 
```

```python
df.head(1)
```

```python
#Qual o custo total?
round(df["Custo"].sum(),2)
```

```python
#Agora que temos a receita e custo e o total, podemos achar o Lucro total
#Vamos criar uma coluna de Lucro que será Receita - Custo
df["lucro"] = df["Valor Venda"] - df["Custo"]
```

```python
df
```

```python
#Criando uma coluna com total de didas para enviar o produto
df["Tempo_envio"] = df["Data Envio"] - df["Data Venda"]
```

```python
df.sample(10)
```

```python
#Extraindo apenas os dias
df["Tempo_envio"] = (df["Data Envio"] - df["Data Venda"]).dt.days
```

```python
df.head(5)
```

```python
df["Tempo_envio"].dtypes
```

```python
#Média do tempo de envio por Marca
df.groupby("Marca")["Tempo_envio"].mean()
```

```python
df.isnull().sum()
```

####**É, se a gente quiser saber o Lucro por Ano e por Marca?**

```python
#Vamos agrupar por ano e marca
df.groupby([df["Data Venda"].dt.year, "Marca"])["lucro"].sum()
```

```python
#corrigir as casas decimais do calculo acima
pd.options.display.float_format = '{:20,.2f}'.format
```

```python
#Alterar o formato colunar com o index
df.groupby([df["Data Venda"].dt.year, "Marca"])["lucro"].sum().reset_index()
```

```python
#Qual o total de produtos vendidos?
df.groupby("Produto")["Quantidade"].sum().sort_values(ascending=False).reset_index()
```

```python
#Qual o total de produtos vendidos?
#Criando gráfico
df.groupby("Produto")["Quantidade"].sum().sort_values(ascending=True).plot.barh(title = "Quantidade de produtos");
plt.xlabel("Quantidade")
plt.ylabel("Produto")
plt.legend();
```

```python
 df.groupby(df["Data Venda"].dt.year)["lucro"].sum().plot.bar(title="Lucro x Ano");
 plt.xlabel("Ano");
 plt.ylabel("Receita");
```

```python
#Selecionar vendas por ano
df.groupby(df["Data Venda"].dt.year)["lucro"].sum().reset_index() 
```

```python
#Selecionar vendas pelo ano de 2019
df_2009 = df[df["Data Venda"].dt.year == 2009]
```

```python
df_2009.head(5)
```

```python
df_2009.groupby(df_2009["Data Venda"].dt.month)["lucro"].sum().plot(title ="Lucro x Mês")
plt.xlabel("Mes")
plt.ylabel("Lucro");
```

```python
#Criando gráfico de Classe x Lucro
df_2009.groupby("Classe")["lucro"].sum().plot.bar(title="Lucro por Marca");
plt.xlabel("Marca")
plt.ylabel("Lucro")
plt.xticks(rotation='horizontal');
```
